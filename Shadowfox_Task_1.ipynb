```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# New Section

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
students = ["A","B","C","D"]
marks = [85, 90, 75, 80]

plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
x = [5, 7, 8, 7, 2, 17, 2, 9]
y = [99, 86, 87, 88, 100, 86, 96, 91]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

```python
sizes = [40, 30, 20, 10]
labels = ['A','B','C','D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```
