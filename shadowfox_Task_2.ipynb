```python
def get_aqi_subindex(concentration, conc_breakpoints, aqi_breakpoints):
    """Calculates the AQI sub-index for a given pollutant concentration."""
    if concentration is None or pd.isna(concentration):
        return None

    if concentration <= conc_breakpoints[0]:
        return aqi_breakpoints[0]
    elif concentration >= conc_breakpoints[-1]:
        return aqi_breakpoints[-1]

    for i in range(len(conc_breakpoints) - 1):
        c_low, c_high = conc_breakpoints[i], conc_breakpoints[i+1]
        i_low, i_high = aqi_breakpoints[i], aqi_breakpoints[i+1]

        if c_low <= concentration <= c_high:
            aqi = ((i_high - i_low) / (c_high - c_low)) * (concentration - c_low) + i_low
            return aqi
    return None


def calculate_pm25_aqi(concentration):
    conc_breakpoints = [0, 12.0, 35.4, 55.4, 150.4, 250.4, 500.4]
    aqi_breakpoints = [0, 50, 100, 150, 200, 300, 500]
    return get_aqi_subindex(concentration, conc_breakpoints, aqi_breakpoints)

def calculate_pm10_aqi(concentration):
    conc_breakpoints = [0, 54, 154, 254, 354, 424, 604]
    aqi_breakpoints = [0, 50, 100, 150, 200, 300, 500]
    return get_aqi_subindex(concentration, conc_breakpoints, aqi_breakpoints)

def calculate_no2_aqi(concentration):

    conc_breakpoints = [0, 53, 100, 360, 649, 1249]
    aqi_breakpoints = [0, 50, 100, 150, 200, 300]

    if concentration > 1249:
         return get_aqi_subindex(concentration, [1249, 2049], [300, 500])
    return get_aqi_subindex(concentration, conc_breakpoints, aqi_breakpoints)


def calculate_so2_aqi(concentration):

    conc_breakpoints = [0, 35, 75, 185, 304, 604]
    aqi_breakpoints = [0, 50, 100, 150, 200, 300]

    if concentration > 604:
         return get_aqi_subindex(concentration, [604, 1004], [300, 500])
    return get_aqi_subindex(concentration, conc_breakpoints, aqi_breakpoints)

def calculate_co_aqi(concentration):

    conc_breakpoints = [0.0, 4.4, 9.4, 12.4, 15.4, 30.4, 40.4, 50.4]
    aqi_breakpoints = [0, 50, 100, 150, 200, 300, 400, 500]
    return get_aqi_subindex(concentration, conc_breakpoints, aqi_breakpoints)

def calculate_o3_aqi(concentration):

    conc_breakpoints_8hr = [0.000, 0.054, 0.070, 0.085, 0.105, 0.200]
    aqi_breakpoints_8hr = [0, 50, 100, 150, 200, 300]

    conc_breakpoints_1hr = [0.201, 0.404, 0.504, 0.604]
    aqi_breakpoints_1hr = [301, 400, 500, 600]
    if concentration > conc_breakpoints_8hr[-1]:

         return get_aqi_subindex(concentration, conc_breakpoints_1hr, aqi_breakpoints_1hr)
    else:
        return get_aqi_subindex(concentration, conc_breakpoints_8hr, aqi_breakpoints_8hr)
```

```python
df['pm2.5_aqi'] = df['pm2.5'].apply(calculate_pm25_aqi)
df['pm10_aqi'] = df['pm10'].apply(calculate_pm10_aqi)
df['no2_aqi'] = df['no2'].apply(calculate_no2_aqi)
df['so2_aqi'] = df['so2'].apply(calculate_so2_aqi)
df['co_aqi'] = df['co'].apply(calculate_co_aqi)
df['o3_aqi'] = df['o3'].apply(calculate_o3_aqi)

df[['pm2.5', 'pm2.5_aqi', 'pm10', 'pm10_aqi', 'no2', 'no2_aqi', 'so2', 'so2_aqi', 'co', 'co_aqi', 'o3', 'o3_aqi']].head()
```

```python
aqi_subindex_cols = ['pm2.5_aqi', 'pm10_aqi', 'no2_aqi', 'so2_aqi', 'co_aqi', 'o3_aqi']
df['AQI'] = df[aqi_subindex_cols].max(axis=1)
display(df[['date'] + aqi_subindex_cols + ['AQI']].head())
```
